```python
#code to segment mp4 files

import subprocess
import os
import math

def get_video_duration(filename):
    result = subprocess.run(["ffprobe", "-v", "error", "-show_entries",
                             "format=duration", "-of",
                             "default=noprint_wrappers=1:nokey=1", filename],
        stdout=subprocess.PIPE,
        stderr=subprocess.STDOUT)
    return float(result.stdout)

def split_video(input_file, segment_length=600):
    # Get the duration of the video
    duration = get_video_duration(input_file)
    
    # Calculate the number of segments
    num_segments = math.ceil(duration / segment_length)
    
    # Get the base name of the input file (without extension)
    base_name = os.path.splitext(input_file)[0]
    
    for i in range(num_segments):
        start_time = i * segment_length
        output_file = f"{base_name}_part{i+1}.mp4"
        
        # FFmpeg command to split the video
        cmd = [
            "ffmpeg",
            "-i", input_file,
            "-ss", str(start_time),
            "-t", str(segment_length),
            "-c", "copy",
            output_file
        ]
        
        # Run the FFmpeg command
        subprocess.run(cmd)
        
        print(f"Created segment: {output_file}")

# Usage
input_file = "test1.mp4"
split_video(input_file)
```

Step 1: Load and Preprocess JSON Files

```python
import os
import json
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Directory containing JSON files
TRANSCRIPTS_FOLDER = "Transcripts"

# Load transcripts and preprocess
def load_transcripts():
    documents = []
    metadata = []
    for file in os.listdir(TRANSCRIPTS_FOLDER):
        if file.endswith(".json"):
            with open(os.path.join(TRANSCRIPTS_FOLDER, file), 'r') as f:
                segments = json.load(f)
                for segment in segments:
                    text = preprocess_text(segment['text'])
                    documents.append(text)
                    metadata.append({
                        "file": file,
                        "start": segment["start"],
                        "end": segment["end"],
                        "original_text": segment["text"]
                    })
    return documents, metadata

# Preprocess text: remove punctuation, lowercase, etc.
def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
    return text

documents, metadata = load_transcripts()
```

Step 2: Build the TF-IDF Index

```python
# Build the TF-IDF vectorizer and fit to the documents
vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform(documents)

# Save the vectorizer and metadata
import pickle
with open("vectorizer.pkl", "wb") as f:
    pickle.dump(vectorizer, f)

with open("metadata.pkl", "wb") as f:
    pickle.dump(metadata, f)
```

Step 3: Query Matching

```python
def search_query(query):
    # Preprocess the query
    query = preprocess_text(query)
    
    # Transform the query into the same vector space
    query_vec = vectorizer.transform([query])
    
    # Compute cosine similarity
    similarity_scores = cosine_similarity(query_vec, tfidf_matrix).flatten()
    top_indices = similarity_scores.argsort()[-5:][::-1]  # Top 5 results
    
    # Fetch metadata for the top results
    results = []
    for idx in top_indices:
        if similarity_scores[idx] > 0:  # Ignore zero similarity scores
            result = metadata[idx]
            result["score"] = similarity_scores[idx]
            results.append(result)
    return results

# Example query
query = "introduction to podcasting"
results = search_query(query)

# Print results
for res in results:
    print(f"File: {res['file']}, Start: {res['start']}, End: {res['end']}, Text: {res['original_text']}, Score: {res['score']:.2f}")
```

```python
from flask import Flask, request, jsonify

app = Flask(__name__)

# Load vectorizer and metadata
with open("vectorizer.pkl", "rb") as f:
    vectorizer = pickle.load(f)

with open("metadata.pkl", "rb") as f:
    metadata = pickle.load(f)

tfidf_matrix = vectorizer.transform([doc['original_text'] for doc in metadata])

@app.route("/search", methods=["POST"])
def search():
    query = request.json.get("query", "")
    results = search_query(query)
    return jsonify(results)

if __name__ == "__main__":
    app.run(debug=True)
```
